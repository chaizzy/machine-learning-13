```python
df["BMI"]=df["BMI"].replace(0,df["BMI"].mean())
df["Pregnancies"]=df["Pregnancies"].replace(0,df["Pregnancies"].mean())
df["Glucose"]=df["Glucose"].replace(0,df["Glucose"].mean())
df["BloodPressure"]=df["BloodPressure"].replace(0,df["BloodPressure"].mean())
df["SkinThickness"]=df["SkinThickness"].replace(0,df["SkinThickness"].mean())
```

```python
df.describe()
```

```python
fig = plt.figure(figsize=(15,10))
sns.boxplot(data=df,width=0.5)
```

```python
# ANS 3 :
x=df.iloc[:,:-1]
y=df["Outcome"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train.shape , y_train.shape
```

```python
# ANS 4:
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred
```

```python
treeclassifier = DecisionTreeClassifier(max_depth=4)
```

```python
treeclassifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
x_train
```

```python
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
parameter = {
    'criterion' : ['gini' , 'entropy' , 'log_loss'],
    'splitter' : ['best' , 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
clf = GridSearchCV(treeclassifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred1 = clf.predict(x_test)
```

```python
y_pred1
```

```python
# ANS 5:
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score = accuracy_score(y_pred1,y_test)
score
```

```python
print(classification_report(y_pred1,y_test))
```

```python
a = confusion_matrix(y_pred1,y_test)
a
```

```python
plt.imshow(a, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
